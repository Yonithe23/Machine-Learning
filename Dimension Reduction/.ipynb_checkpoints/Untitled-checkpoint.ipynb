```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('6.1 train.csv')
```

```python
data.head()
```

```python
X = data.iloc[:,:-1]
Y = data.iloc[:, -1]
```

# Data Standrdnizing 

```python
mean = X.mean(axis=0)
sig = X.std(axis=0)
```

```python
X = (X-mean)/sig
```

```python
X.head()
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
best_features = SelectKBest()
```

```python
fit = best_features.fit(X, Y)
fit.scores_
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
features_Scores = pd.concat([dfcolumns , dfscores], axis = 1)
```
